```python
import pandas as pd
```

```python
StormEvents = pd.read_csv('Downloads/StormEvents_details-ftp_v1.0_d2016_c20180718.csv.gz')
SevereWeatherDatabase = pd.read_csv('Downloads/2016_torn.csv')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
StormEvents = StormEvents[StormEvents['EVENT_TYPE'] == 'Tornado']
plt.figure(figsize = (11*3/4, 8.5*3/4))
plt.scatter(y = StormEvents['BEGIN_LAT'], x = StormEvents['BEGIN_LON'],
            color = 'yellow', marker = 'o', s=75)
plt.scatter(y = SevereWeatherDatabase['slat'], x = SevereWeatherDatabase['slon'],
            color = 'navy', marker = '.', s= 50);
```

```python
f, axarr = plt.subplots(1,2)
f.set_figheight(8.5)
f.set_figwidth(22)

a = axarr[0].hist2d(y = StormEvents['BEGIN_LAT'], 
                    x = StormEvents['BEGIN_LON'],
                    bins = 50, cmap = 'inferno_r')
b = axarr[1].hist2d(y = SevereWeatherDatabase['slat'],
                    x = SevereWeatherDatabase['slon'],
                    bins = 50, cmap = 'inferno_r');
```

```python
plt.hist((a[0] - b[0]).ravel());
```
